```python
import pandas as pd
```

```python
## Loading in a sample of real time MI solar farm production data from https://www.nrel.gov/grid/solar-power-data
solar_sample = pd.read_csv('Actual_42.55_-85.35_2006_DPV_15MW_5_Min.csv')

## grabbing average daily solar output data from current season
solar_sample['LocalTime'] = pd.to_datetime(solar_sample['LocalTime'], format='%m/%d/%y %H:%M')

# Assume df has a datetime column called "time" and a forecast column "solar_output"
solar_sample.set_index("LocalTime", inplace=True)

# Resample to hourly, taking the mean (or sum, max, etc. depending on your data)
solar_sample = solar_sample.resample("H").sum()
```

```python
train_solar = solar_sample.loc['2006-01-01 00:00:00':'2006-11-27 00:00:00', :]
test_solar = solar_sample.loc['2006-11-27 01:00:00':'2006-12-31 23:00:00', :]
```

```python
## grabbing historical cloud cover data, will be used to adjust solar models to account for any abnormal cloud coverage over the sample period
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
import requests

import requests
import pandas as pd

latitude = 42.55
longitude = -85.35
start = "2006-01-01"
end = "2006-12-31"
timezone = "America/New_York"

url = (
    f"https://archive-api.open-meteo.com/v1/archive?"
    f"latitude={latitude}&longitude={longitude}"
    f"&start_date={start}&end_date={end}"
    f"&hourly=cloudcover,temperature_2m,wind_speed_10m"
    f"&timezone={timezone}"
)

response = requests.get(url)
data = response.json()

# Convert to DataFrame
df = pd.DataFrame(data["hourly"])

# Hourly datetime data
df["time"] = pd.to_datetime(df["time"], format="%Y-%m-%dT%H:%M")

# Assume df has a datetime column called "time" and a forecast column "solar_output"
df.set_index("time", inplace=True)
```

```python
## converting into cloud cover over prediction and cloud cover forecast
train_features = df.loc['2006-01-01 00:00:00':'2006-11-27 00:00:00', :]
test_features = df.loc['2006-11-27 01:00:00':'2006-12-31 23:00:00', :]
```

```python
import matplotlib.pyplot as plt
# Method 1: Using matplotlib for multiple lines on the same plot
plt.figure(figsize=(12, 6))
plt.plot(train_solar.index, train_solar['Power(MW)'], label='Power(MW)', linewidth=2)
plt.plot(train_features.index, train_features['cloudcover'], label='cloudcover', alpha=0.7)
plt.plot(train_features.index, train_features['temperature_2m'], label='temperature_2m', alpha=0.7)
plt.plot(train_features.index, train_features['wind_speed_10m'], label='wind_speed_10m', alpha=0.7)
plt.title('Multiple Time Series Features vs Target Variable')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.deterministic import CalendarFourier, DeterministicProcess
import pandas as pd
import matplotlib.pyplot as plt

# Make sure the index is datetime and set frequency
train_solar = train_solar.asfreq('H')  # Set daily frequency

# Create Fourier seasonal terms
fourier = CalendarFourier(freq='D', order=1)  # daily seasonality, 3 harmonics
dp = DeterministicProcess(index=train_solar.index, constant=True,
                          order=1,
                          seasonal=False,
                          additional_terms=[fourier],
                          drop=True)

X = dp.in_sample()
X = pd.concat([X, train_features], axis = 1)
model = SARIMAX(train_solar['Power(MW)'], order=(1, 0, 1), exog=X)
result = model.fit()
print(result.summary())
```

```python
# predict and evaluate
n_forecast = len(test_features)
X_future = dp.out_of_sample(steps=n_forecast)
X_future = pd.concat([X_future, test_features], axis = 1)
                     
# Using 'result' model
forecast = result.get_forecast(steps=n_forecast, exog=X_future)

# Convert the forecast object to a pandas DataFrame first
forecast_df = forecast.conf_int()

# Add the predicted mean as a column
forecast_df["predicted_mean"] = forecast.predicted_mean

# Now clip the values
forecast_df["predicted_mean"] = forecast_df["predicted_mean"].clip(lower=0)

#RMSE
rmse = mean_squared_error(test_solar, forecast_df["predicted_mean"], squared=False)
print(f"Test RMSE: {rmse:.3f}")
```
